```python
import numpy as np
import pandas as pd
```

```python
df_saudesono = pd.read_csv('saudesono_vida.csv')
df_saudesono.head()
```

```python
#Questão 1
df_saudesono.rename(columns={'ID': 'Identificador', 'Pressão sanguíneaaaa': 'Pressão sanguínea', 'Ocupação': 'Profissão', 'Categoria BMI': 'Categoria IMC'}, inplace=True)
df_saudesono.head()
```

```python
media_sono = df_saudesono.groupby('Profissão')['Duração do sono'].mean()
mediana_sono = df_saudesono.groupby('Profissão')['Duração do sono'].median()

moda_sono = df_saudesono.groupby('Profissão')['Duração do sono'].agg(pd.Series.mode)

print("\nMédia da duração de sono por profissões:")
print(media_sono)

print("\nMediana da duração de sono por profissões:")
print(mediana_sono)

print("\nModa da duração de sono por profissões:")
print(moda_sono)
```

```python
result = df_saudesono.groupby('Profissão')['Duração do sono'].agg(['mean', 'median', pd.Series.mode])
result.columns = ['Média', 'Mediana', 'Moda']
result
```

```python
percentual_sobrepeso = (
    df_saudesono[df_saudesono['Profissão'] == 'Eng. de Software']['Categoria IMC']
    .eq('Sobrepeso')
    .mean() * 100
)

print(f"Porcentagem de pessoas com sobrepeso entre os Engenheiros de Software: {percentual_sobrepeso:.2f}%")
```

```python
#Questão 4
quest_pr = ['Representante de Vendas','Advogado(a)']
```

```python
media_sono_interesse = df_saudesono[df_saudesono['Profissão'].isin(quest_pr)].groupby('Profissão')['Duração do sono'].mean()
```

```python
media_sono_outros = df_saudesono[~df_saudesono['Profissão'].isin(quest_pr)]['Duração do sono'].mean()
```

```python
profissao_menor_sono = media_sono_interesse.idxmin()
menor_sono = media_sono_interesse.min()
```

```python
print(f"\nProfissão que faz dormir menos: {profissao_menor_sono}")
print(f"Média de duração do sono para {profissao_menor_sono}: {menor_sono:.2f} horas")
```

```python
if media_sono_interesse.mean() < media_sono_outros:
    print("Profissões de Advogado(a) ou Representante de Vendas fazem você dormir menos comparado ao resto.")
else:
    print("Profissões de Advogado(a) ou Representante de Vendas não fazem você dormir menos comparado ao resto.")
```

```python
df_mulheres = df_saudesono[df_saudesono['Gênero'] == 'Mulher']
df_homens = df_saudesono[df_saudesono['Gênero'] == 'Homem']
df_saudesono['Pressão Float'] = df_saudesono['Pressão sanguínea'].str.replace('/', '.').astype(float)


mediapr_mulheres = df_mulheres['Pressão Float'].mean()
mediapr_homens = df_homens['Pressão Float'].mean()
print(f"Média da pressão sanguínea das mulheres: {mediapr_mulheres:.2f}")
print(f"Média da pressão sanguínea dos homens: {mediapr_homens:.2f}")
if mediapr_mulheres > mediapr_homens:
    print("As mulheres têm uma pressão sanguínea média maior do que os homens.")
else:
    print("Os homens têm uma pressão sanguínea média maior do que as mulheres.")
```

```python
df_acima_70 = df_saudesono[df_saudesono['Frequência cardíaca'] > 70]
df_abaixo_ou_igual_70 = df_saudesono[df_saudesono['Frequência cardíaca'] <= 70]
```

```python
media_passos_70 = df_acima_70['Passos diários'].mean()
media_passos_ab = df_abaixo_ou_igual_70['Passos diários'].mean()

print(f"Média de passos diários com frequência cardíaca acima de 70: {media_passos_70:.2f}")
print(f"Média de passos diários com frequência cardíaca abaixo ou igual a 70: {media_passos_ab:.2f}")
if media_passos_70 > media_passos_ab:
    print("Participantes com frequência cardíaca acima de 70 têm em média mais passos diários do que os com frequência cardíaca abaixo ou igual a 70.")
else:
    print("Participantes com frequência cardíaca abaixo ou igual a 70 têm em média mais passos diários do que os com frequência cardíaca acima de 70")
```
