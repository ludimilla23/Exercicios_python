```python
def separar_pressao(pressao):
    sistolica, diastolica = pressao.split('/')
    return float(sistolica), float(diastolica)

df_mulheres = df_saudesono[df_saudesono['Gênero'] == 'Mulher'].copy()
df_homens = df_saudesono[df_saudesono['Gênero'] == 'Homem'].copy()

df_mulheres[['Pressão Sistólica', 'Pressão Diastólica']] = df_mulheres['Pressão sanguínea'].apply(lambda x: pd.Series(separar_pressao(x)))
df_mulheres['PAM'] = (df_mulheres['Pressão Sistólica'] + 2 * df_mulheres['Pressão Diastólica']) / 3

df_homens[['Pressão Sistólica', 'Pressão Diastólica']] = df_homens['Pressão sanguínea'].apply(lambda x: pd.Series(separar_pressao(x)))
df_homens['PAM'] = (df_homens['Pressão Sistólica'] + 2 * df_homens['Pressão Diastólica']) / 3

media_pam_mulheres = df_mulheres['PAM'].mean()
media_pam_homens = df_homens['PAM'].mean()

if media_pam_mulheres > media_pam_homens:
    print(f"As mulheres têm uma pressão arterial média maior que os homens. (Média PAM das mulheres: {media_pam_mulheres:.2f} mmHg, Média PAM dos homens: {media_pam_homens:.2f} mmHg)")
else:
    print(f"Os homens têm uma pressão arterial média maior que as mulheres. (Média PAM dos homens: {media_pam_homens:.2f} mmHg, Média PAM das mulheres: {media_pam_mulheres:.2f} mmHg)")


```

```python
df_acima_70 = df_saudesono[df_saudesono['Frequência cardíaca'] > 70]
df_abaixo_ou_igual_70 = df_saudesono[df_saudesono['Frequência cardíaca'] <= 70]
```

```python
media_passos_70 = df_acima_70['Passos diários'].mean()
media_passos_ab = df_abaixo_ou_igual_70['Passos diários'].mean()

print(f"Média de passos diários com frequência cardíaca acima de 70: {media_passos_70:.2f}")
print(f"Média de passos diários com frequência cardíaca abaixo ou igual a 70: {media_passos_ab:.2f}")
if media_passos_70 > media_passos_ab:
    print("Participantes com frequência cardíaca acima de 70 têm em média mais passos diários do que os com frequência cardíaca abaixo ou igual a 70.")
else:
    print("Participantes com frequência cardíaca abaixo ou igual a 70 têm em média mais passos diários do que os com frequência cardíaca acima de 70")
```
